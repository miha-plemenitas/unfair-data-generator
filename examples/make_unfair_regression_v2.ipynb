# Unfair Regression Dataset Generation and Analysis

This tutorial demonstrates **`unfair_data_generator` extended for regression**.
The regression generator reuses the unfair classification data structure
(features and sensitive groups) and generates a **group-dependent continuous target**.

Unlike standard regression datasets, this approach enables **controlled unfairness**
in regression tasks, such as systematic group bias and heteroscedastic noise.

The notebook systematically explores how **different regression base functions**
interact with **different unfairness mechanisms** and how these effects are reflected
in group-wise fairness metrics.


## Overview

This notebook covers:
- **Dataset generation** for unfair regression
- **Base regression functions** (linear and non-linear)
- **Unfairness mechanisms** in regression
- **Model training** using a fixed regression model
- **Fairness analysis** using group-wise regression metrics

### Key Functions Covered
1. `make_unfair_regression` – generates unfair regression datasets
2. `train_and_evaluate_regression` – trains a model and computes group metrics
3. `visualize_regression_metrics` – visualizes fairness disparities


```python
# Required Imports

import sys
import os

sys.path.append(os.path.abspath('..'))

import numpy as np
import pandas as pd

from unfair_data_generator.unfair_regression import make_unfair_regression
from unfair_data_generator.util.regression_trainer import train_and_evaluate_regression
from unfair_data_generator.util.regression_visualizer import visualize_regression_metrics
```

## 1. Experimental Configuration

We evaluate unfair regression across **two independent dimensions**:

### Regression base functions
These define the global relationship between features and the target:
- **linear** – linear combination of informative features
- **logistic** – bounded, non-linear transformation
- **exponential** – non-linear transformation with exponential growth

### Regression unfairness types
These define how unfairness is introduced across sensitive groups:
- **Equal MSE** – equal noise and no systematic bias (fair setting)
- **Group bias** – systematic over-/underestimation for some groups
- **Heteroscedastic noise** – group-dependent noise levels

All combinations of base functions and unfairness types are evaluated.


```python
# Global experiment configuration

n_samples = 3000
n_features = 10
n_informative = 3
n_sensitive_groups = 3
random_state = 42

base_functions = ["linear", "logistic", "exponential"]
fairness_types = ["Equal MSE", "Group bias", "Heteroscedastic noise"]
```

## 2. Running the Full Experiment Grid

For each combination of:
- regression base function, and
- regression unfairness type,

we:
1. generate an unfair regression dataset,
2. train the same regression model,
3. compute group-wise fairness metrics (MSE, MAE, Bias),
4. store the results for later comparison.


```python
# Run experiments for all combinations

results = {}

for base_fn in base_functions:
    results[base_fn] = {}

    for fairness_type in fairness_types:
        X, y, Z = make_unfair_regression(
            n_samples=n_samples,
            n_features=n_features,
            n_informative=n_informative,
            fairness_type=fairness_type,
            base_function=base_fn,
            n_sensitive_groups=n_sensitive_groups,
            random_state=random_state
        )

        metrics = train_and_evaluate_regression(X, y, Z)
        results[base_fn][fairness_type] = metrics
```

## 3. Inspecting a Representative Dataset

Before analyzing metrics, we inspect one representative dataset
to understand the structure of the generated data.

Below we examine:
- base function: **linear**
- unfairness type: **Group bias**


```python
# Inspect one example dataset

X, y, Z = make_unfair_regression(
    n_samples=n_samples,
    n_features=n_features,
    n_informative=n_informative,
    fairness_type="Group bias",
    base_function="linear",
    n_sensitive_groups=n_sensitive_groups,
    random_state=random_state
)

df = pd.DataFrame(X, columns=[f"x{i}" for i in range(X.shape[1])])
df["y"] = y
df["Z"] = Z

df.head()
```

## 4. Group-wise Target Distribution

A sanity check:
If unfairness is present, different sensitive groups should exhibit
different target distributions.


```python
df.groupby("Z")["y"].describe()
```

## 5. Numerical Comparison of Regression Fairness Metrics

In addition to visual inspection, we now present a **numerical comparison**
of fairness metrics.

For each combination of:
- regression base function, and
- unfairness type,

we report **group-wise values** of:
- Mean Squared Error (MSE)
- Mean Absolute Error (MAE)
- Bias

These tables allow precise, quantitative comparison and support the
interpretation of the visual plots shown later.


```python
# Numerical comparison tables

summary_tables = {}

for base_fn in base_functions:
    summary_tables[base_fn] = {}

    for fairness_type in fairness_types:
        metrics = results[base_fn][fairness_type]

        rows = []
        for group, m in metrics.items():
            rows.append({
                "Group": group,
                "MSE": round(m["MSE"], 4),
                "MAE": round(m["MAE"], 4),
                "Bias": round(m["Bias"], 4),
            })

        df_metrics = pd.DataFrame(rows)
        summary_tables[base_fn][fairness_type] = df_metrics

        print("=" * 70)
        print(f"Base function: {base_fn} | Fairness type: {fairness_type}")
        print("=" * 70)
        display(df_metrics)
```

## 6. Visual Comparison Across All Settings

We now visualize regression fairness metrics for **all combinations**
of base regression functions and unfairness types.

Each plot shows, per sensitive group:
- Mean Squared Error (MSE)
- Bias (mean signed error)

This allows direct comparison of how different unfairness mechanisms
manifest in regression.


```python
for base_fn in base_functions:
    for fairness_type in fairness_types:
        title = f"Unfair Regression – {base_fn} base | {fairness_type}"
        visualize_regression_metrics(results[base_fn][fairness_type], title)
```

## 7. Interpretation of Results

- **Equal MSE**
  - Similar MSE and MAE across groups
  - Bias values close to zero
  - Represents a fair regression scenario

- **Group bias**
  - Clear differences in Bias between groups
  - Indicates systematic over- or underestimation

- **Heteroscedastic noise**
  - Differences in MSE and MAE across groups
  - Bias remains close to zero
  - Some groups are inherently harder to predict

The **base regression function** controls the global shape of the target,
while the **unfairness type** determines how disparities arise across groups.
